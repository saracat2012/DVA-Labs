```python
import pandas as pd

df = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1')

df.head()
```

```python
df['bpm'].plot(kind='hist')

```

```python
df['bpm'].describe()
```

```python
df['mode'].value_counts().plot(kind='pie')
```

```python
df['mode'].value_counts()
```

```python

filtered_df = df.loc[(df['bpm'] >= 85) & (df['bpm'] <= 140) & (df['mode'] == 'Minor') & (df['key'] == 'A') ]

filtered_df.head()
```

```python
df['key_mode'] = df['key'] + ' ' + df['mode']
df['key_mode'].value_counts()

```

```python
cols = ['track_name', 'artist(s)_name', 'in_spotify_charts', 
        'danceability_%', 'energy_%', 'speechiness_%', 'acousticness_%']
subset = df[cols].copy()

subset = subset.dropna(subset=['track_name', 'artist(s)_name'])

numeric_cols = ['in_spotify_charts', 'danceability_%', 'energy_%', 'speechiness_%', 'acousticness_%']
subset[numeric_cols] = subset[numeric_cols].apply(pd.to_numeric, errors='coerce')

sorted_by_charts = subset.sort_values(by='in_spotify_charts', ascending=False)

print(sorted_by_charts.head(100))

means = sorted_by_charts[numeric_cols].mean()
print(means)
```

```python
import matplotlib.pyplot as plt

numeric_cols = ['in_spotify_charts', 'danceability_%', 'energy_%', 'speechiness_%', 'acousticness_%']

for col in numeric_cols:
    plt.figure(figsize=(8,5))
    sorted_by_charts[col].hist(bins=30, edgecolor='black')  # pandas histogram
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```
